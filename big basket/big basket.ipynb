```python
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
import bs4
from urllib.request import urlopen
from bs4 import BeautifulSoup

from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(r'D:\chromedriver_win32\\chromedriver')
driver.get('https://www.bigbasket.com/sp/2008045_pooja-needs/?nc=Season%20Must%20Haves&t_pg=Aug-T1-Home-Page&t_p=Aug-T1-2020&t_s=Season%20Must%20Haves&t_pos_sec=4&t_pos_item=4&t_ch=desktop#!page=2')
sleep(7)
```

```python
page_source = driver.page_source
sleep(4)

soup = BeautifulSoup(page_source, 'lxml')
type(soup)

bs4.BeautifulSoup

```

```python
title = soup.title
print(title)
```

```python
list_rows1 = []
for strong_tag in soup.find_all("div", {"class":"col-sm-12 col-xs-7 prod-name"}):
    print(strong_tag.text)
    list_rows1.append(strong_tag.text)
```

```python
list_rows2 = []
for strong_tag in soup.find_all("div", {"class":"col-sm-12 col-xs-7 qnty-selection"}):
    print(strong_tag.text)
    list_rows2.append(strong_tag.text)
```

```python
list_rows3 = []
for strong_tag in soup.find_all("div", {"class":"col-sm-12 col-xs-12 add-bskt"}):
    print(strong_tag.text)
    list_rows3.append(strong_tag.text)
```

```python
df = pd.DataFrame(data={"Product Name":list_rows1,"Product-qunty":list_rows2,"Product-cost":list_rows3})
df.to_csv(r"D:\BIGBASKET2.csv", sep=',',index=False)
df.head(50)
```
